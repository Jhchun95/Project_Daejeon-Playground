```python
# 대전공원정보 전처리 파일
import pandas as pd
```

```python
# 데이터 불러오기

# 대덕구 변수
daeduckgu_park = pd.read_csv('G:/bigdata/daeduckgu_park.csv', encoding = 'cp949')
# 동구 변수
donggu_park = pd.read_csv('G:/bigdata/donggu_park.csv',encoding ='cp949')
# 중구 변수
junggu_park = pd.read_csv('G:/bigdata/junggu_park.csv',encoding ='cp949')
# 서구 변수
seogu_park = pd.read_csv('G:/bigdata/seogu_park.csv',encoding ='cp949')
# 유성구 변수
yuseonggu_park = pd.read_csv('G:/bigdata/yuseong_park.csv',encoding ='cp949')
```

```python
# 대전 동구 공원의 컬럼을 호출함.
donggu_park.columns
```

```python
# 대전 대덕구 공원의 컬럼을 호출함.
daeduckgu_park.columns
```

```python
# 동구와 대덕구 등 정보를 불러와봤는데, 변수명이 모두 다 같은 것을 확인할 수 있었음.
# 변수명이 다 같으므로 daejeon_park라는 새 변수를 선언하고, 행 밑으로 결합함.
daejeon_park = pd.concat([daeduckgu_park, donggu_park, junggu_park, seogu_park, yuseonggu_park])
```

```python
# daejeon_park의 처음 3줄의 데이터를 출력함.
# head : DataFrame 내의 처음 n줄의 데이터를 출력함.
daejeon_park.head(3)
```

```python
# 행과 열 개수를 가져옴.
print(daeduckgu_park.shape,
donggu_park.shape,
junggu_park.shape,
seogu_park.shape,
yuseonggu_park.shape) 
```

```python
# shape : 축 정보 확인함.
daejeon_park.shape
```

```python
# 대전광역시 어느 구인지 변수를 추가하기 위해서 변수를 확인함.
daejeon_park['소재지지번주소'].head() 
```

```python
# 빈 리스트 생성 
gu = [] 

for gu1 in daejeon_park['소재지지번주소']: 
    # ex> 대전광역시 대덕구 중리동388-17의 5번째부터 9번쨰 글자를 가져옴.
    gu1 = gu1[5:9] 
    # 공백을 공백없음으로 바꿔주고, " 대덕구" " 중구 " 형식으로 가져옴.
    # replace 함수는 공백을 제거하기 위해서 사용함.
    gu1 = gu1.replace(" ","") 
    # 뽑아온 문자열을 gu이라는 리스트에 저장함.  
    gu.append(gu1) 
```

```python
# 전처리한 gu 변수를 확인함.
gu
```

```python
# 동 변수 추가함. (맨마지막열)
daejeon_park['gu'] = gu 
```

```python
# 대전 공원의 첫 2줄을 출력함.
daejeon_park.head(2)
```

```python
# 전처리한 데이터프레임을 엑셀 csv로 저장함. 
daejeon_park.to_csv('G:/bigdata/daejeon_park.csv', encoding ='cp949')
```
